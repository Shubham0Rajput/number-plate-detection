```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plate = cv2.imread('DATA/car_plate.jpg')
plate = cv2.cvtColor(plate,cv2.COLOR_BGR2RGB)
plt.imshow(plate)
```

```python
plate_cascade= cv2.CascadeClassifier('DATA/haarcascade_plate_number.xml')
```

```python
plate_img=plate.copy()
rec=plate_cascade.detectMultiScale(plate_img,scaleFactor=1.2,minNeighbors=5)
for (x,y,w,h) in rec:
    cv2.rectangle(plate_img,(x,y),(x+w,y+h),(255,0,0),5)
plt.imshow(plate_img)
```
